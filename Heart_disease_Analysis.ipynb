```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os


file_path  = project_path+"Heart_failure_datasets.csv"
pd.read_csv(file_path)
print("data loaded sucessfully")
```

```python
df.dropna(inplace = True)
df = df[(df["RestingBP"] > 0) & (df["Cholesterol"] > 0) & (df["MaxHR"] > 0)]
print("Data cleaned ,Ready for Analysis")
```

```python
import pandas as pd
import io

data_string = """Age,Sex,ChestPainType,RestingBP,Cholesterol,FastingBS,RestingECG,MaxHR,ExerciseAngina,Oldpeak,ST_Slope,HeartDisease
0,40,M,ATA,140,289,0,Normal,172,N,0.0,Up,0
1,49,F,NAP,160,180,0,Normal,156,N,1.0,Flat,1
2,37,M,ATA,130,283,0,ST,98,N,0.0,Up,0
3,48,F,ASY,138,214,0,Normal,108,Y,1.5,Flat,1
4,54,M,NAP,150,195,0,Normal,122,N,0.0,Up,0
"""
grouped_df = df.groupby(['Sex', 'HeartDisease'])[['Age', 'RestingBP', 'Cholesterol', 'MaxHR']].mean()
```

```python
charts_path = os.path.join(project_path, "charts/")
os.makedirs(charts_path, exist_ok=True)
```

```python
#visulization
import plotly.express as px

# Reset the index to make 'Sex' and 'HeartDisease' regular columns for plotting
plot_df = grouped_df.reset_index()
```

```python
#visualization 1
fig_age = px.bar(
    plot_df,
    x='Sex',
    y='Age',
    color='HeartDisease',
    barmode='group',
    title='Mean Age by Sex and Heart Disease',
    labels={'HeartDisease': 'Heart Disease (0=No, 1=Yes)'},
    hover_data={'Sex': True, 'HeartDisease': True, 'Age': ':.2f'}
)
plt.savefig(charts_path + "Mean_age.png")
plt.close()
```

```python
fig_bp = px.bar(
    plot_df,
    x='Sex',
    y='RestingBP',
    color='HeartDisease',
    barmode='group',
    title='Mean Resting Blood Pressure by Sex and Heart Disease',
    labels={'HeartDisease': 'Heart Disease (0=No, 1=Yes)'},
    hover_data={'Sex': True, 'HeartDisease': True, 'RestingBP': ':.2f'}
)
plt.savefig(charts_path + "Resting_BP.png")
plt.close()
```

```python
fig_cholesterol = px.bar(
    plot_df,
    x='Sex',
    y='Cholesterol',
    color='HeartDisease',
    barmode='group',
    title='Mean Cholesterol by Sex and Heart Disease',
    labels={'HeartDisease': 'Heart Disease (0=No, 1=Yes)'},
    hover_data={'Sex': True, 'HeartDisease': True, 'Cholesterol': ':.2f'}
)
plt.savefig(charts_path + "Mean_Cholestrol.png")
plt.close()
```

```python
fig_maxhr = px.bar(
    plot_df,
    x='Sex',
    y='MaxHR',
    color='HeartDisease',
    barmode='group',
    title='Mean Max Heart Rate by Sex and Heart Disease',
    labels={'HeartDisease': 'Heart Disease (0=No, 1=Yes)'},
    hover_data={'Sex': True, 'HeartDisease': True, 'MaxHR': ':.2f'}
)
plt.savefig(charts_path + "Heart_Rate.png")
plt.close()



print(f"✅ Charts saved to folder: {charts_path}")

# Export cleaned data
cleaned_path = project_path + "cleaned_heart_disease.csv"
df.to_csv(cleaned_path, index=False)
print(f"✅ Cleaned data exported to: {cleaned_path}")
```
